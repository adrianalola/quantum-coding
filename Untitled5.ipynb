```python
# === Todo en una sola celda ===
# Inicializa Q# con perfil compatible
import qsharp
qsharp.init(target_profile=qsharp.TargetProfile.Base)

# Compila una operación Q# mínima (sin namespace, compatible con qsharp.compile)
from qsharp import compile as qcompile
RandomNBits = qcompile("""
    open Microsoft.Quantum.Intrinsic;
    open Microsoft.Quantum.Measurement;

    operation RandomNBits(n : Int) : Result[] {
        mutable results = [];
        for _ in 1..n {
            use q = Qubit();
            H(q);
            let r = M(q);
            Reset(q);
            set results += [r];
        }
        return results;
    }
""")

# Simula 1000 tiros de 1 bit
results = [str(RandomNBits.simulate(n=1)[0]) for _ in range(1000)]
zeros, ones = results.count("Zero"), results.count("One")

# Grafica y guarda PNG
import matplotlib.pyplot as plt
plt.bar(["Zero","One"], [zeros, ones])
plt.title("Distribución de Random Bits (1000 tiros)")
plt.xlabel("Resultado"); plt.ylabel("Frecuencia")
plt.tight_layout()
plt.savefig("random_bits_hist.png", dpi=150)
plt.show()

print("Listo ✅  Zero:", zeros, " One:", ones)
```
